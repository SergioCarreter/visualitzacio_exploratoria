* Visualizacion de 3 variables: 2 numericas y una categorica

Consideremos los datos de los vuelos con retraso _**X_late**_

Grafiquemos 'ArrDelay' vs 'DepDelay' agrupado por 'UniqueCarrier'

Para ello usaremos el dataframe agrupado _**grouped**_

```python
plt.figure(figsize=(12,12))
leg=[]
k=0
colors=['lightcoral', 'violet', 'purple', 'gray', 'cyan', 'lightgray', 'green', 'greenyellow', 'limegreen',
        'yellow', 'darkgoldenrod','darkorange', 'goldenrod', 'lightsalmon', 'magenta', 'maroon', 'darkmagenta',
        'black', 'blue','red']
for name, group in grouped:
    leg.append(name)
    plt.scatter(group['DepDelay'], group['ArrDelay'], c=colors[k])
    k=k+1
plt.legend(leg)
plt.xlabel('DepDelay')
plt.ylabel('ArrDelay')
plt.title('ArrDelay vs DepDelay agrupado por UniqueCarrier', fontsize=25)
plt.savefig('arrdelay_depdelay_carrier.png', bbox_inches='tight')
plt.show()
```

Visualicemos la relacion entre 'AirTime' y 'DepDelay' en los vuelos con retraso (matriz X_late)

```python
airtime_l = X_late['AirTime']
print(airtime_l.min(), ';', airtime_l.count(), airtime_l.shape)
```

```python
plt.figure(figsize=(9,9))
plt.scatter(depdelay_l, airtime_l)
plt.axis('equal')
plt.title('Airtime vs DepDelay', fontsize=20)
plt.xlabel('DepDelay [min]')
plt.ylabel('AirTime [min]')
plt.savefig('airtime_depdelay.png', bbox_inches='tight')
plt.show()
```

```python
# Establezco tamaño del grafico
plt.rcParams['figure.figsize']=(12,12)

# Guardo en ax los 3 ejes que ademas seran en 3D
ax= plt.axes(projection='3d')

#Doy contenido a los ejes (x,y,z)
ax.scatter3D(depdelay_l, airtime_l, arrdelay_l )

#Pongo labels a los ejes
ax.set_xlabel('DepDelay')
ax.set_ylabel('AirTime')
ax.set_zlabel('ArrDelay')

# Giro a conveniencia (elevacion, azimutal) en grados (0,0) -> (z,y); (90,0) ->(y,x)
ax.view_init(45,205)

plt.title('ArrDelay = f(DepDelay,AirTime)', fontsize=20)
plt.savefig('arrdelay_depdelay_airtime.png', bbox_inches='tight')
plt.show()

#type(ax)
```

* Visualizacion de mas de 3 variables: 3 numericas y una categorica.

```python
# Establezco tamaño del grafico
plt.rcParams['figure.figsize']=(12,12)

# Guardo en ax los 3 ejes que ademas seran en 3D
ax= plt.axes(projection='3d')

leg=[]
k=0
colors=['lightcoral', 'violet', 'purple', 'gray', 'cyan', 'lightgray', 'green', 'greenyellow', 'limegreen',
        'yellow', 'darkgoldenrod','darkorange', 'goldenrod', 'lightsalmon', 'magenta', 'maroon', 'darkmagenta',
        'black', 'blue','red']
for name, group in grouped:
    leg.append(name)
    ax.scatter3D(group['DepDelay'], group['AirTime'], group['ArrDelay'], c=colors[k] )
    k=k+1
    
#Aplico la leyenda
plt.legend(leg)

#Pongo labels a los ejes
ax.set_xlabel('DepDelay')
ax.set_ylabel('AirTime')
ax.set_zlabel('ArrDelay')

# Giro a conveniencia (elevacion, azimutal) en grados (0,0) -> (z,y); (90,0) ->(y,x)
ax.view_init(25,205)

plt.title("ArrDelay = f(DepDelay,AirTime) agrupado por 'UniqueCarrier'", fontsize=20)
plt.savefig('4variables3.png', bbox_inches='tight')
plt.show()
```
